Detecting the missing values

```python
#checking for missing values
wines.isnull().sum()
```

```python
#imputing the missing values by means as the data present in the different columns are continuous values
for col in wines.columns:
    if wines[col].isnull().sum()>0:
        wines[col]=wines[col].fillna(wines[col].mean())
wines.isnull().sum().sum()
```

Data Preprocessing

Data Preprocessing is the process of evaluating , filtering , manipulating , and encoding data so that a Machine Learning Algorithm can understand it and use the resulting output . The main goal of Data Preprocessing is to eliminate data issues such as missing values , improve data quality , and make the data useful for  Machine Learning purposes .

```python
#separating the data and label
X=wines.drop('quality',axis=1)
print(X)
```

Label Binarization

Here , we are converting the categorical values into the binary values i.e., 0 & 1 is said to be as Linear Binarization .

```python
Y=wines['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
print(Y)
```

Train_Test_Split Method

train_test_split method is a technique that divides a Dataset into two subsets one is a training set & a test set . And is generally used in Supervised Machine Learning for Regression & Classification of model .

1 . Train set : Training set is used to train the model . 

2 . Test set : The Test set is used to evaluate the model's performance .

While we are training and testing the sets it contains the text_size ( describes the number that defines the size of the test set i.e., here we taken as 25%(0.25)) and 
random_state ( is a parameter that controls the random number generator used to shuffle the data before splitting it ) within the paranthesis .

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=3)
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

Model Training :

Random Forest Classifier

A Random Forest Classifier is a Machine Learning Algorithm that combines the output of multiple decision trees to reach a single result . 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

Accuracy is a metric that measures how often a Machine Learning Model correctly predicts the output . We can calculate accuracy as : 
                       
Accuracy = Number of Correct Predictions divided by Total Number of Predictions .
                        

```python
from sklearn.metrics import accuracy_score
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_wines_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy :',test_wines_accuracy)
```

 Here we are converting the input data into numpy array because numpy arrays are a powerful tool for Machine Learning . By converting our data into numpy arrays , we can improve the memory efficiency , speed , and capability of our code . 

 And also we are reshaping the data that helps the input data more useful and managable . It makes the data easy to accessible , analyse and more informative as by reshaping we can convert or transform data from one format to another format .

 As we declared above that the y- value is 1 if the quality is greater than 7 and y-value is 0 if quality is below 7  . Based on this , we are going to get the output i.e., quality .

Building a Predictive System

```python
given_data=(7.8,0.58,0.02,2,0.073,9,18,0.9968,3.36,0.57,9.5)

#converting the given data into a numpy array
numpy_array=np.asarray(given_data)

#as we are predicting the label for only one instance so we're reshaping the data
reshaped_data=numpy_array.reshape(1,-1)

prediction=model.predict(reshaped_data)
print(prediction)

if (prediction[0]==1):
    print('Wine Quality is Good...')
else:
    print('Wine Quality is Not Good...')
```
